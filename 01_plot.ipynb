```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import csv
import util
from pprint import pprint
```

```python
trainfile = 'data/train.csv'
testfile = 'data/test.csv'

(header, ids, X, Y) = util.fetch_data('data/train.csv', impute=True)
m, n = X.shape

X_pos = X[Y==1,:]
X_neg = X[Y==0, :]
```

```python
npairs = 20
pairs = sorted([sorted(np.random.choice(n, 2, replace=False)) for _ in range(npairs)])
```

```python
for (a, b) in pairs:
    plt.figure()
    plt.scatter(X_pos[:,a], X_pos[:,b], color='r', marker='^')
    plt.scatter(X_neg[:,a], X_neg[:,b], color='b', marker='v')
    plt.xlabel('x_{}'.format(a))
    plt.ylabel('x_{}'.format(b))
    plt.savefig('figures/eda_plots/{}_{}.png'.format(a,b), format='png', dpi=100)
```
