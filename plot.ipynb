```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import csv
import util
from pprint import pprint
```

```python
trainfile = 'data/train.csv'
testfile = 'data/test.csv'

(header, X, Y) = util.fetch_data('data/train.csv')
m, n = X.shape

X_pos = X[Y==1,:]
X_neg = X[Y==0, :]
```

```python
npairs = 100
pairs = [np.random.choice(57, 2, replace=False) for _ in range(npairs)]
```

```python
for (a, b) in pairs:
    plt.figure()
    plt.scatter(X_pos[:,a], X_pos[:,b], color='r', marker='^')
    plt.scatter(X_neg[:,a], X_neg[:,b], color='b', marker='v')
    plt.xlabel('x_{}'.format(a))
    plt.ylabel('x_{}'.format(b))
    plt.show()
```
